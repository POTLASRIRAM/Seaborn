### Plotting relationships using Seaborn point plot

```python
sns.catplot('Product Category','Shipping Cost',kind='point',data=df)
```

```python
sns.catplot('Product Category','Shipping Cost',kind='point',hue='Ship Mode',data=df)
```

```python
sns.catplot('Product Category','Shipping Cost',kind='point',hue='Ship Mode',hue_order=['Express Air','Delivery Truck','Regular Air'],data=df)
```

```python
sns.catplot('Product Category','Shipping Cost',kind='point',row='Ship Mode',hue_order=['Express Air','Delivery Truck','Regular Air'],data=df)
```

```python
sns.catplot('Product Category','Shipping Cost',kind='point',col='Ship Mode',hue_order=['Express Air','Delivery Truck','Regular Air'],data=df)
```

```python
sns.catplot('Product Category','Shipping Cost',kind='point',col='Ship Mode',row='Customer Segment',hue_order=['Express Air','Delivery Truck','Regular Air'],data=df)
```
