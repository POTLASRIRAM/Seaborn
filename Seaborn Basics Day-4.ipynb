### Plotting relationships using Seaborn violin plot

```python
sns.catplot('Product Category','Shipping Cost',kind='violin',data=df)
```

```python
sns.catplot('Product Category','Shipping Cost',kind='violin',hue='Ship Mode',data=df)
```

```python
sns.catplot('Product Category','Shipping Cost',kind='violin',hue='Ship Mode',hue_order=['Express Air','Delivery Truck','Regular Air'],data=df)
```

```python
sns.catplot('Product Category','Shipping Cost',kind='violin',row='Ship Mode',hue_order=['Express Air','Delivery Truck','Regular Air'],data=df)
```

```python
sns.catplot('Product Category','Shipping Cost',kind='violin',col='Ship Mode',hue_order=['Express Air','Delivery Truck','Regular Air'],data=df)
```

```python
sns.catplot('Product Category','Shipping Cost',kind='violin',col='Ship Mode',row='Customer Segment',hue_order=['Express Air','Delivery Truck','Regular Air'],data=df)
```
